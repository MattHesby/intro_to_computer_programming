```python
import requests
import lxml.html
from bs4 import BeautifulSoup
```

```python
url = 'https://nuevaschool.instructure.com'
username = 'cshan'
password = open('./password.txt').read().strip()
```

```python
login_url = url + '/login/canvas'
```

```python
s = requests.Session()
login = s.get(login_url)
login_html = lxml.html.fromstring(login.text)
hidden_inputs = login_html.xpath(r'//form//input[@type="hidden"]')
form = {x.attrib["name"]: x.attrib["value"] for x in hidden_inputs}
```

```python
form['pseudonym_session[unique_id]'] = username
form['pseudonym_session[password]'] = password
```

```python
form
```

```python
response = s.post(login_url, data=form)
```
